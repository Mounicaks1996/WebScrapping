```python
hotelname = [] 
address=[]
rating=[]
prices=[]
facility=[]
driver=webdriver.Chrome(ChromeDriverManager().install())
for i in range(1,15):
    start_time = time.time()

    URL = 'https://www.oyorooms.com/hotels-in-bangalore/?page='+str(i)
    driver.get(URL)
    content=driver.page_source
    soup=BeautifulSoup(content)
    list1=soup.find_all("a", attrs={"class":"c-nn640c u-width100"})
    list2=soup.find_all("div", attrs={"class":"d-body-lg listingHotelDescription__hotelAddress"})
    list4=soup.find_all("div", attrs={"class":"hotelRating"})
    list5=soup.find_all("span", attrs={"class":"listingPrice__finalPrice"})
    list6=soup.find_all("div", attrs={"class":"amenityWrapper"})
    for i in list1:
        #print(type(i))
        #print(i.text) 
        if i is None:
            hotelname.append(np.NaN)
        else:
            hotelname.append(i.text)
            
    for j in list2:            
        if j is None:
            address.append(np.NaN)
        else:
            address.append(j.text)
            
    for s in list4:       
        if s is None:
            rating.append(np.NaN)
        else:
            rating.append(s.text)
            
    for t in list5:       
        if t is None: 
            prices.append(np.NaN)
        else:
            prices.append(t.text)
            
    for a in list6: 
        if a is None:
            facility.append(np.NaN)
        else:
            facility.append(a.text)
            
            
            
            
            
            
            
            
            
```
